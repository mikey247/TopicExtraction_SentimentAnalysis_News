```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
from pandas import read_csv
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk
```

```python
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
# Create a BERTopic model
topic_model = BERTopic(verbose=True, embedding_model=embedding_model)

# abc_news = read_csv("./abcnews_sample.csv")
# docs = abc_news["headline_text"].tolist()

bbc_news = read_csv("./bbc_news.csv")
docs = bbc_news["description"].tolist()
# docs = [""]
```

```python
# Fit the model on the documents
topics, _ = topic_model.fit_transform(docs)

info = topic_model.get_topic_info()
print(info)

# Get the most frequent topic
topic_freq = topic_model.get_topic_freq()
print(topic_freq)

# Save the model
topic_model.save("models/my_model")
```
